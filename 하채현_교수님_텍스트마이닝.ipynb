```python
import pandas as pd
file_path = "/content/drive/MyDrive/한류와 한국어 전공 베트남 학생들의 한국문화 인식(설문응답) (1).xlsx"

data = pd.read_excel(file_path, sheet_name = "Sheet1")
data = data[2:].drop('Unnamed: 0', axis = 1).reset_index(drop = True)
data
```

```python
import re

def remove_special_characters(input_string):
    clean_string = re.sub(r'[^\w\s]', ' ', input_string)
    return clean_string
```

```python
#full text
insang_ft = remove_special_characters(" ".join([ word.upper() for word in data.iloc[:,1].tolist() ])).replace("  ", " ")
gyounghum_ft = remove_special_characters(" ".join([ word.upper() for word in data.iloc[:,3].tolist() ])).replace("  ", " ")
il_ft = remove_special_characters(" ".join([ word.upper() for word in data.iloc[:,4].tolist() ])).replace("  ", " ")
```

```python
for en, kor in trans_list:
  insang_ft = insang_ft.replace(en, kor)
  gyounghum_ft = gyounghum_ft.replace(en, kor)
  il_ft = il_ft.replace(en, kor)

insang_ft
```

```python
#형태소 분석기
from konlpy.tag import Okt

okt = Okt()
insang_morphs = okt.nouns(insang_ft)
gyounghum_morphs = okt.nouns(gyounghum_ft)
il_morphs = okt.nouns(il_ft)
```

```python
def 블핑변환기(word):
  old_word = "블핑"
  new_word = "블랙핑크"

  return [new_word if word == old_word else word for word in word]

def 케이팝변환기(word):
  old_word = "팝"
  new_word = "케이팝"

  return [new_word if word == old_word else word for word in word]

def 케이드라마변환기(word):
  old_word = "드라마"
  new_word = "케이드라마"

  return [new_word if word == old_word else word for word in word]

def 한글자제거기(my_list):
  new_list = [word for word in my_list if len(word) > 1]

  return new_list

def 빈도분석(word):
  morphs = 블핑변환기(word)
  morphs = 케이팝변환기(morphs)
  morphs = 케이드라마변환기(morphs)
  morphs = 한글자제거기(morphs)

  from collections import Counter

  word_count = Counter(morphs)

  cnt_list = []
  total_count = len(morphs)

  for word, count in word_count.items():
      cnt_dict = {}
      cnt_dict['word'] = word
      cnt_dict['N'] = count
      cnt_dict['%'] = round((count / total_count) * 100, 2)

      cnt_list.append(cnt_dict)

  cnt_df = pd.DataFrame(cnt_list).sort_values('N', ascending = False).iloc[:20,].reset_index(drop=True)

  return cnt_df
```

```python
빈도분석(insang_morphs).to_csv('인상.csv', encoding = 'cp949')
```

```python
# 불필요한 단어 제거
def 단어제거기(my_list, del_word):
  new_list = [word for word in my_list if word != del_word]

  return new_list

il_morphs = 단어제거기(il_morphs,'관련')
il_morphs = 단어제거기(il_morphs,'기업')
il_morphs = 단어제거기(il_morphs,'회사')
il_morphs = 단어제거기(il_morphs,'분야')
il_morphs = 단어제거기(il_morphs,'커뮤')
il_morphs = 단어제거기(il_morphs,'한국')

빈도분석(il_morphs).to_csv('일.csv', encoding = 'cp949')
```

```python
#워드클라우드용
def calculate_word_frequency(morphs):

    word_count = {}
    for noun in morphs:
        if noun not in word_count:
            word_count[noun] = 1
        else:
            word_count[noun] += 1

    return word_count

# 텍스트 데이터
insang_count = calculate_word_frequency(insang_morphs)
gyounghum_count = calculate_word_frequency(gyounghum_morphs)
il_count = calculate_word_frequency(il_morphs)
```
