```python
file_path = "/content/drive/MyDrive/20230808 전채은님 신뢰도 분석용.xlsx"
```

```python
import pandas as pd

data_so = pd.read_excel(file_path, sheet_name = '소').dropna()
data_daeji = pd.read_excel(file_path, sheet_name = '돼지').dropna()
data_dak = pd.read_excel(file_path, sheet_name = '닭').dropna()
```

```python
so_columns_list = data_so['문항명'].unique().tolist()
daeji_columns_list = data_daeji['문항명'].unique().tolist()
dak_columns_list = data_dak['문항명'].unique().tolist()
```

```python
from pyirr import kappam_fleiss

temp_list = []
temp_columns = [(data_so, so_columns_list), (data_daeji, daeji_columns_list), (data_dak, dak_columns_list)]

for i, j in temp_columns:
  for k in range(len(j)):
    temp_dict = {}
    temp_data = i[i['문항명'] == j[k]]

    temp_dict['가축'] = temp_data['가축'].unique()[0]
    temp_dict['문항번호'] = temp_data['문항'].unique()[0]
    temp_dict['문항명'] = temp_data['문항명'].unique()[0]
    temp_dict['Fleiss` Kappa'] = round(kappam_fleiss(temp_data.iloc[:, -3:]).value, 3)

    temp_list.append(temp_dict)

temp_df = pd.DataFrame(temp_list)
temp_df
```
